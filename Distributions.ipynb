```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
gammavar = np.random.gamma(8,1,100)
print(np.mean(gammavar))
print(np.std(gammavar))
plt.hist(gammavar, edgecolor='k', alpha=0.5)
plt.axvline(gammavar.mean(), color='k', linestyle='dashed', linewidth=1)
plt.axvline(gammavar.std(), color='k', linestyle='solid', linewidth=1)
plt.show()
```
